```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.tail()
```

```python
print(iris.groupby('species').size())
```

### Distribution plot(Rugplot,displotandHistplot)

```python
sns.rugplot(x='petal_width',y='petal_length',data=iris,height=0.2)
```

```python
sns.displot(data=iris,x='petal_length')
```

```python
sns.displot(data=iris,x='petal_length',kde=True,hue='petal_length')
```

```python
sns.displot(data=iris,x='petal_length',rug=True,kde=True,rug_kws={"height": 0.5})
```

```python
sns.histplot(data=iris,x='petal_length',bins=20)
```

```python
sns.set(style='white')
sns.histplot(data=iris,x='petal_length',bins=20)
```

```python
iris.hist(edgecolor='green',linewidth=0.8,figsize=(18,8))
plt.show()
```

### Categorical Plots(Boxplot,Countplot,Barplot,Violinplot)

```python
data=np.arange(1,21,1)
sns.boxplot(data=data)
```

```python
sns.boxplot(data=iris,x='species',y='petal_length',hue='species')
```

```python
sns.boxplot(data=iris,x='species',y='petal_length',hue='species',width=0.5)
```

# Countplot

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='sepal_length',data=iris)
plt.show()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='sepal_length',data=iris,hue='species')
plt.show()
```

```python
 plt.figure(figsize=(10,4),dpi=200)
sns.countplot(x='sepal_length',data=iris,hue='species',palette='Paired')
plt.show()
```

### Barplot

```python
plt.figure(figsize=(10,6),dpi=200)
sns.barplot(x='sepal_length',y='sepal_width',data=iris,errorbar='sd',estimator=np.mean)
```

```python
plt.figure(figsize=(10,6),dpi=200)
sns.barplot(x='sepal_length',y='sepal_width',data=iris,errorbar='sd',estimator=np.mean,hue='species')
```

```python
plt.figure(figsize=(10,6),dpi=200)
sns.barplot(x='sepal_length',y='sepal_width',data=iris,errorbar='sd',estimator=np.mean,hue='species')
plt.legend(loc='upper right')
```

### Violin Plot

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
sns.violinplot(x='species',y='sepal_length',data=iris)
plt.subplot(2,2,2)
sns.violinplot(x='species',y='sepal_width',data=iris)
plt.subplot(2,2,3)
sns.violinplot(x='species',y='petal_length',data=iris)
plt.subplot(2,2,4)
sns.violinplot(x='species',y='petal_width',data=iris)
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='species',y='sepal_length',data=iris,hue='species')
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='species',y='sepal_length',data=iris,hue='species',split=True)
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='species',y='sepal_length',data=iris,inner=None)
```

```python
plt.figure(figsize=(20,10))
sns.violinplot(x='species',y='sepal_length',data=iris,inner='quartile')
```
